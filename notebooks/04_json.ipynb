4. Écrire une boucle pour récupérer les informations de toutes les planètes disponibles dans l'API et stocker le résultat dans un dataframe `planets`.

```python
url_base = "https://swapi-node.vercel.app"
url_page = "/api/planets"

planets_dfs = []
while url_page is not None:
        url = url_base + url_page
        print(f"Téléchargement de {url}")
        r = requests.get(url)
        r.raise_for_status

        planets = r.json()
        planets_dfs.append(
            pd.DataFrame(
                 [planet["fields"] for planet in planets["results"]]
            )
        )

        url_page = planets["next"]

planets_df = pd.concat(planets_dfs, ignore_index=True)
# mettre à jour l'url de la page
# sur cette page, récupérer le dictionnaire fields qui est dans la partie  results en dataframe
# concatener au dataframe précédent
```

5. Exporter le dataframe obtenu à la question précédente dans un fichier `planets.json` au format NDJSON.

```python
planets_df.to_json("planets.json", orient="records", lines=True)
```

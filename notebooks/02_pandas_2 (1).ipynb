8. Extraire la liste des participants à partir des colonnes `Player1` et `Player2`. Une façon de faire consiste à utiliser `concat` et la méthode `drop_duplicates` pour obtenir le résultat sous la forme d'une série et de la convertir en dataframe avec la méthode `to_frame`.

```python

rg_joueur = 
(
    pd.concat([tennis["Player1"], tennis["Player2"]])
    .drop_duplicates()   # récupérer les valeurs uniques
    .to_frame(name="Joueur")  # convertit série en dataframe
    .reset_index(drop=True)  # reset index vec drop = false : ramène la colonne en index en colonne, # reset index avec drop = true : supprime lancienne colonne d'index
)
rg_joueur
```

9. Écrire une fonction `n_match` qui prend une chaîne de caractères `joueur` en entrée et retourne le nombre de matchs disputés par le joueur.

```python
def n_match(joueur):
    return len(tennis.loc[(tennis["Player1"] == joueur) | (tennis["Player2"] == joueur)])

rg_joueur["n_match"]=rg_joueur["Joueur"].apply(n_match)
rg_joueur
```

16. Construire un dataframe `rg_victoires` avec les trois colonnes suivantes pour le tournoi de Roland Garros :
- `joueur` : nom du joueur tel qu'il est donné dans `rg`,
- `nom_joueur` : nom de famille du joueur uniquement,
- `n_victoire` : nombre de matchs gagnés dans le tournoi.

```python
# étape 1 : fair eavec 1 seul joueur
joueur = "Roger Federer"
n_victoire1 = len(rg[(rg["Player1"] == joueur) & (rg["Result"] == 1)]) 
n_victoire2 = len(rg[(rg["Player2"] == joueur) & (rg["Result"] == 0)])
n_victoire = n_victoire1 + n_victoire2
n_victoire
```

```python
# étape 2 : généraliser dans une fonction
def n_victoires(df, joueur):
    n_victoire1 = len(df[(df["Player1"] == joueur) & (df["Result"] == 1)]) 
    n_victoire2 = len(df[(df["Player2"] == joueur) & (df["Result"] == 0)])
    n_victoire = n_victoire1 + n_victoire2
    return n_victoire
```

```python
# étape 3 : appliquer fonction
n_victoires(rg, "Roger Federer")
    
```

```python
# étape 4 : intégration dans une variable
rg_victoires = (
    pd.concat([rg["Player1"], rg["Player2"]])
    .drop_duplicates()   # récupérer les valeurs uniques
    .to_frame(name="joueur")  # convertit série en dataframe
    .reset_index(drop=True)  # reset index vec drop = false : ramène la colonne en index en colonne, # reset index avec drop = true : supprime lancienne colonne d'index
    )

rg_victoires["nom_joueur"] = (
    rg_victoires["joueur"]
    .str.split(" ")
    .apply(lambda v: v[-1])
    )

rg_victoires["n_victoire"] = (
    rg_victoires["joueur"]
    .apply(lambda joueur: n_victoires(rg, joueur))
    )
```

17. Construire un dataframe `wb_victoires` avec les trois colonnes suivantes pour le tournoi de Wimbledon :
- `joueur` : nom du joueur tel qu'il est donné dans `wb`,
- `nom_joueur` : nom de famille du joueur uniquement,
- `n_victoire` : nombre de matchs gagnés dans le tournoi.

```python
wb_victoires = (
    pd.concat([wb["Player1"], wb["Player2"]])
    .drop_duplicates()   # récupérer les valeurs uniques
    .to_frame(name="joueur")  # convertit série en dataframe
    .reset_index(drop=True)  # reset index vec drop = false : ramène la colonne en index en colonne, # reset index avec drop = true : supprime lancienne colonne d'index
)

wb_victoires["nom_joueur"] = (
    wb_victoires["joueur"]
    .str.split(".") # différence de'écriture de nom dans wb
    .apply(lambda v: v[-1])
    )

wb_victoires["n_victoire"] = (
    wb_victoires["joueur"]
    .apply(lambda joueur: n_victoires(wb, joueur))
)
```

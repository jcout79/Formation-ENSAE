7. Calculer la moyenne et la variance de la variable `SepalLength` pour chaque espèce.

```python
df[["SepalLength"]].groupby(df.Species).agg(["mean", "var"])
```

8. Calculer la moyenne de toutes les variables pour chaque espèce. Afficher le résultat au format large et au format long.

```python
moy = df[["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]].groupby(df.Species).mean()
#transformation format long
print(
    moy.reset_index()
    .melt(id_vars = ["Species"], value_name = "measure")
)
```
